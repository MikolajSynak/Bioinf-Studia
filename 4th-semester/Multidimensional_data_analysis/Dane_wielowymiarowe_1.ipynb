###### 4. Zamień pierwszą kolumnę z trzecią kolumną, a drugi wiersz z ostatnim.

```python
# Switch the positions of the first and third columns
data = data.iloc[:, [2, 1, 0] + list(range(3, len(data.columns)))]

# Swap the positions of the second and last rows
data = data.iloc[[len(data)-1] + list(range(1, len(data)-1)) + [0], :]

data
```
