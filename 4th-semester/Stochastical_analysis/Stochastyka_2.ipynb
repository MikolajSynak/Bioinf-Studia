```python
import numpy as np
import matplotlib.pyplot as plt

Mrx = np.array([[0, 0.5, 0.5, 0, 0], 
                [0.5, 0, 0, 0.5, 0], 
                [0, 0, 0.7, 0.3, 0], 
                [0, 0, 0.5, 0.5, 0], 
                [1, 0, 0, 0, 0]])

p0 = [1, 0, 0, 0, 0]

num_steps = 10

# Initialize empty lists to store the probabilities for each state at each step
prob0_list = []
prob1_list = []
prob2_list = []
prob3_list = []
prob4_list = []

# Iterate over each step and compute the probabilities for each state
for i in range(num_steps):
    prob = np.linalg.matrix_power(Mrx, i)
    prob0 = np.dot(p0, prob)
    prob1 = prob[1][2]
    prob2 = prob[4][0]
    prob3 = prob[2][3]
    prob4 = prob[3][3]
    prob0_list.append(prob0)
    prob1_list.append(prob1)
    prob2_list.append(prob2)
    prob3_list.append(prob3)
    prob4_list.append(prob4)

# Plot the probabilities for each state over time

plt.plot(range(num_steps), prob1_list, label='State 1')
plt.plot(range(num_steps), prob2_list, label='State 2')
plt.plot(range(num_steps), prob3_list, label='State 3')
plt.plot(range(num_steps), prob4_list, label='State 4')
plt.xlabel('Step')
plt.ylabel('Probability')
plt.title('Probability of Being in Each State over Time')
plt.legend()
plt.show()
```
