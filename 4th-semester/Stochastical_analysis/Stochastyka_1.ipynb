```python
lista1 = []

def machina1(x):
    while x < 101:
        a = random.choice(sequence)
        lista1.append(a)
        x = x + 1

machina1(1)

plt.plot(lista1)
plt.show()

srednia1 = np.mean(lista1)
srednia1 = round(srednia1, 4)
print(f'Średnia = {srednia1}')
```
