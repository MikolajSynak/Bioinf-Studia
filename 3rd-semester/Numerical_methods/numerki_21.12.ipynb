```python
import numpy as np
from math import pi

# Define sampling frequency
fs = 1000
dt = 1/fs

# Define time arrays for each note duration
t_whole = np.arange(0, 4, dt)
t_half = np.arange(0, 2, dt)
t_quarter = np.arange(0, 1, dt)
t_eighth = np.arange(0, 0.5, dt)
t_sixteenth = np.arange(0, 0.25, dt)

# Define frequencies for each note
A3 = 220
C4 = 261.63
D4 = 293.66
E4 = 329.63
G4 = 392

# Define time signature and tempo (in beats per minute)
time_sig = 7
bpm = 126

# Define sine waves for each note
A3_whole = np.sin(2*pi*A3*t_whole)
C4_quarter = np.sin(2*pi*C4*t_quarter)
D4_quarter = np.sin(2*pi*D4*t_quarter)
E4_quarter = np.sin(2*pi*E4*t_quarter)
G4_quarter = np.sin(2*pi*G4*t_quarter)

C4_eighth = np.sin(2*pi*C4*t_eighth)
D4_eighth = np.sin(2*pi*D4*t_eighth)
E4_eighth = np.sin(2*pi*E4*t_eighth)
G4_eighth = np.sin(2*pi*G4*t_eighth)

C4_sixteenth = np.sin(2*pi*C4*t_sixteenth)
D4_sixteenth = np.sin(2*pi*D4*t_sixteenth)
E4_sixteenth = np.sin(2*pi*E4*t_sixteenth)
G4_sixteenth = np.sin(2*pi*G4*t_sixteenth)

# Define full song as an array by concatenating the individual notes
song = np.concatenate((A3_whole, C4_quarter, D4_quarter, E4_quarter,
                       G4_quarter, C4_eighth, D4_eighth, E4_eighth,
                       G4_eighth, C4_sixteenth, D4_sixteenth, E4_sixteenth,
                       G4_sixteenth))

# Determine the duration of the song in seconds
duration = len(song) / fs

# Define time array for the full song
t = np.linspace(0, duration, len(song))

import sounddevice as sd

sd.play(song, fs)
```

```python
import sounddevice as sd
import numpy as np

# Set the duration of each note
duration = 0.5  # seconds

# Set the sampling rate and the waveform parameters
sampling_rate = 44100
amplitude = 0.5

# Create a waveform for each note in the scale
waveforms = []
for octave in range(1, 9):  # 8 octaves on a piano
    for note in ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']:  # 12 notes per octave
        if note == 'C':
            frequency = octave * 262
        elif note == 'C#':
            frequency = octave * 277
        elif note == 'D':
            frequency = octave * 294
        elif note == 'D#':
            frequency = octave * 311
        elif note == 'E':
            frequency = octave * 330
        elif note == 'F':
            frequency = octave * 349
        elif note == 'F#':
            frequency = octave * 370
        elif note == 'G':
            frequency = octave * 392
        elif note == 'G#':
            frequency = octave * 415
        elif note == 'A':
            frequency = octave * 440
        elif note == 'A#':
            frequency = octave * 466
        elif note == 'B':
            frequency = octave * 494
        else:
            print(f"Error: Invalid note {note}")
            continue
        waveform = amplitude * np.sin(2 * np.pi * np.arange(sampling_rate * duration) * frequency / sampling_rate)
        waveforms.append(waveform)
print(waveforms)
```

```python
import sounddevice as sd
import numpy as np
import scipy

# Define the frequency of each note
notes = {'E': 329.63, 'A': 440, 'B': 493.88, 'C': 523.25, 'D': 587.33, 'G#': 739.99}

# Create an array of the notes in the melody
melody = ['E', 'A', 'B', 'C', 'A', 'B', 'G#', 'A', 'E', 'A', 'B', 'C', 'D', 'E', 'D']

# Create an array of the duration of each note
durations = [0.5] * len(melody)

# Define the sample rate and the amplitude
fs = 44100
amplitude = 0.5

# Initialize an empty array to store the samples
samples = np.array([])

# Iterate over the notes and durations, generating samples for each one
for note, duration in zip(melody, durations):
    frequency = notes[note]
    t = np.linspace(0, duration, int(fs * duration))
    samples = np.concatenate((samples, amplitude * np.sin(2 * np.pi * frequency * t)))

# Play the samples
sd.play(samples, fs)

f, t, Sxx = scipy.signal.spectrogram(abs(melody))
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
