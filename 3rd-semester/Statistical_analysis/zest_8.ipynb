```python
import math as mt
from scipy import stats
#hipoteza zerowa: mi=mi0
#hipoteza alternatywna: mi=!mi0
mi=122 #srednia populacji
sigma=4 #odchylenie populacji
n=25
mi0=124.6 # sredniaproby
s=5 #odchylenie proby
t=(mi0-mi)/s*mt.sqrt(n)
print(t)
alpha=0.05
```

```python
wk1=stats.t.ppf(1-alpha/2,n-1)
wk2=stats.t.ppf(alpha/2,n-1)
print(wk1,'\n',wk2)
#obszar krytyczny jest (-∞,-2.06] i [2.06,∞)
#t wynosi 2.59, wpada w obszar wiec hipoteze zerowa odrzucamy
#odpowiedz jest 'tak'
```

```python
#odchylenie wagi - grupa pierwsza
s1=24.4
n1=12
#odchylenie wagi - grupa druga
s2=12.4
n2=5
#hipoteza h0-> wariancja1=wariancja2
#hipoteza hA-> wariancja1=!wariancja2
alpha=0.05
#wartosc F
f=(s1**2)/(s2**2)
print(f)
#stopnie swobody
ss1=n1-1
ss2=n2-1
#wartosci krytyczne wyznaczamy z rozkladu F
wk1=stats.f.ppf(1-alpha/2,ss1,ss2)
wk2=stats.f.ppf(alpha/2,ss1,ss2)
print(wk1,'\n',wk2)
x=stats.f.cdf(f,ss1,ss2)
p=2*min(x,1-x)
print(p)
```

# zadanie 8

```python
import math as mt
from scipy import stats
#h0: p1=p2
#hA: p1=!p2
alpha=0.05
n1=102
x1=29
n2=102
x2=34
#test dla frakcji to bedzie
phat1=x1/n1
phat2=x2/n2
print(phat1)
print(phat2)
pline=(x1+x2)/(n1+n2)
sigma=mt.sqrt(pline*(1-pline)*(1/n1+1/n2))
#wartosc testowa
z=(phat1-phat2)/sigma
print(z)
#wartosc krytyczna
wk=stats.norm.ppf(1-alpha/2)
print(wk)
#obszar krytyczny jest (-∞,-1.96] i [1.96,∞]
#z nie wpada w obszar wiec nie odrzucamy h0
p=2*(1-stats.norm.cdf(abs(z)))
print(p)
```

```python
import scipy.stats as stats
#hipoteza h0-> wszystkie wartosci maja takie samo prawdopodobienstwo(1/6)
#hipoteza hA-> 'kostka jest nieuczciwa'
#test zgodnosci chi-kwadrat
obs=[14,15,27,13,22,29]
n=120
#rozklad teoretyczny, robisz n*p, tutaj 120*1/6
exp=[20,20,20,20,20,20]
print(stats.chisquare(obs,exp))
#pvalue mniejsze niz alpha(0.05) wiec odrzucamy h0
#wartosc krytyczna jest z tablic i wynosi 11.07
#obszar krytycczny [11.07,∞)
```
