```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
import statistics as st
from math import ceil
import statsmodels.api as sm
```

```python
srednia=st.mean(x)
print(srednia)
```

```python
mediana=st.median(x)
print(mediana)
```

```python
#odchylenie standardowe
s=st.stdev(x)
print(s)
```

```python
#odpowiednia liczba klas
y=list(range(35,85,10))
```

```python
#rysowanie histogramu
plt.hist(x,bins=y,rwidth=0.85,density=1,cumulative=1)
print(max(x))
print(min(x))
```

```python
#rysowanie histogramu z polem rownym 1
plt.hist(x,bins=y,rwidth=0.85, density=1)
print(max(x))
print(min(x))
```

```python
#dystrybuanta, kumulacja wszystkich binow po kolei(skumulowany rozklad czestosci)
plt.hist(x,bins=y,rwidth=0.85,density=1,cumulative=1)
print(max(x))
print(min(x))
```

```python
#wyznaczanie wartosci podstawowych statystyk
c=st.mean(x) #srednia
d=st.stdev(x) #odchhylenie standardowe
e=st.median(x) #mediana
f=st.mode(x) #moda
print(c,d,e,f)
```

```python
waga=pd.read_excel(r'Waga.xlsx')
print(waga)
```

```python
#kwantyl dolny
q1=waga['Waga'].quantile(0.25)
print(q1)
```

```python
#mediana
q2=waga['Waga'].quantile(0.5)
print(q2)
```

```python
#kwantyl gorny
q3=waga['Waga'].quantile(0.75)
print(q3)
```

```python
#odchylenie standardowe
odchylenie=waga['Waga'].std()
print(odchylenie)
```

```python
#srednia
srednia=waga['Waga'].mean()
print(srednia)
```

```python
#moda
moda=waga['Waga'].mode()
print(moda)
```

```python
#rozstep
roz=waga['Waga'].max()-waga['Waga'].min()
print(roz)
```

```python
#wspolczynnik zmiennosci sredniej
zms=odchylenie/srednia
print(zms)
```

```python
#rozstep kwartylowy
rk=q3-q1
print(rk)
```
