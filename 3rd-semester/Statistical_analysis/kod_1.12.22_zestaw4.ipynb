```python
dane3 = pd.read_excel(r'Rozkłady.xlsx')
print(dane3)
```

```python
# Średnia i mediana dla proby 1:
wzrost = dane3['Proba 1']

Średnia = wzrost.mean()
Mediana = wzrost.median()
Skośność = wzrost.skew()

print('Średnia =', Średnia, '\nMediana =', Mediana, '\nSkośność =', Skośność)
```

##### Histogramy:

```python
#bins = liczba klas (domyślnie jest 10)

plt.hist(wzrost, bins = 12)
plt.axvline(Średnia, color='red', label='Średnia')
plt.axvline(Mediana, color='yellow', label='Mediana')
plt.legend()
plt.xlabel('Wzrost')
plt.show()
```

```python
dlugosc_zycia = dane3['Proba 3']
ŚredniaD = dlugosc_zycia.mean()
MedianaD = dlugosc_zycia.median()
SkośnośćD = dlugosc_zycia.skew()
print(f'Średnia = {ŚredniaD}, \nMediana = {MedianaD}, \nSkośność = {SkośnośćD}')
```

```python
plt.hist(dlugosc_zycia, bins = 12)
plt.axvline(ŚredniaD, color = 'red', label = "Średnia")
plt.axvline(MedianaD, color = 'yellow', label = 'Mediana')
plt.legend()
plt.xlabel('Dlugość życia')
plt.show()
```
